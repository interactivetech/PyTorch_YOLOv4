# Training a YOLOv4 model on the MS COCO dataset

This notebook demonstrates training an 8-bit YOLOv4 network on the MS COCO dataset. The task involves the following steps:
   
1.    Train full precision network.
3.    Measure accuracy of trained full precision network on test set.
2.    Fine-tune 8-bit network.
4.    Measure accuracy of 8-bit network on test set.
5.    Troubleshoot commonly encountered errors.

All code has standard PyTorch syntax except blocks noted as NP-PyTorch-specific.


```python
import logging
import yaml
from tqdm import tqdm
from utils.plots import plot_images, plot_labels, plot_results, plot_evolution
from jupyter_training_utils import train, test
from utils.general import increment_path
from pathlib import Path
import torch

import os

# Set the environment variable
os.environ['CUDA_VISIBLE_DEVICES'] = '0'
os.environ['OMP_NUM_THREADS'] = '4'
#import sys
#sys.path.append('/workspace/np-pytorch/src')
```

## Train a full precision network on MS COCO

### Set run parameters and train network

The specific parameters that are used for the run are specified as follows. They can be changed to modify the run parameters. For example, you can specify a different path to a pretrained checkpoint from which to load weights by changing <code>opt['weights']</code> below or change the training hyper-parameters for a different dataset.

In order to fine-tune the model on a custom dataset, <code>opt['data']</code> should point to the yaml file that contains paths to the train, validation and test datasets, the number of classes, and class names. 

Based on the number of classes in the dataset, update the last conv layers before the YOLO layers and the number of classes in the YOLO layers in <code>opt['cfg']</code> to update the model architecture.

The XVIEW notebook demonstrates an example of fine-tuning the model on an overhead satellite imagery dataset.

```python
opt = {}
opt['weights'] = 'weights/BestCOCO.pt'
opt['cfg'] = 'cfg/yolov4-csp-leaky.cfg'
opt['data'] = 'data/coco.yaml'
opt['hyp'] = 'data/hyp.scratch.LR0.0000625.yaml'
opt['epochs'] = 1
opt['batch_size'] = 2
opt['img_size'] = [640, 640]
opt['resume'] = False #resume previous training run
opt['restart'] = True #restart the most recent training, starting with epoch 0 for the lr_scheduler
opt['notest']= False
opt['workers'] =  8
opt['project'] = 'runs/train'
opt['name'] = 'expLR0.0000625_batch2_1epoch_640x640_coco'
opt['precision'] = 8
opt['conf_thres'] = 0.05
opt['quantizer'] = 'faq'
opt['img_size'].extend([opt['img_size'][-1]] * (2 - len(opt['img_size'])))  # extend to 2 sizes (train, test)
opt['save_dir'] = increment_path(Path(opt['project']) / opt['name'], exist_ok=False)  # increment run

FPlogger = logging.getLogger(__name__)
```

The following code trains a model initiliazed using the cfg file above in 'unconstrained' mode in NP-Pytorch, which allows training a full precision model, equivalent to native pytorch. Passing <code>train_mode='unconstrained'</code> as an argument to the train function below ensures this behavior.

Details of the full train function are omitted here. The train function can be found in <code>jupyter_training_utils.py</code>.

For the purpose of demonstration, the weights are initialized using trained weights from a 300 epoch training run on MS COCO, and the notebook trains for 2 additional epochs. In practice, it is recommended to train a full precision model for 300 or more epochs on a new dataset. Increasing or lowering the initial learning rate helps imporove results. When using a larger (or smaller) batch size, scale up (or down) the learning rate linearly.

```python
RUN_TRAINING = False

if RUN_TRAINING:
    #Use GPU for training if available
    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

    # Load Hyperparameters from file
    with open(opt['hyp']) as f:
        hyp = yaml.load(f, Loader=yaml.FullLoader)  # load hyps

    # Train in unconstrained mode
    FPlogger.info(opt)
    results = train(hyp, opt, device, FPlogger, train_mode='unconstrained')
```

## Measure task accuracy of trained full precision network on test set

Measure task accuracy of the released trained model weights, or model weights saved in the previous step on the test set below. 

To measure task accuracy on a different model, change <code>opt['weights']</code> to point to trained model weights below or change the test hyper-parameters for a different dataset.

```python
opt = {}
opt['weights'] = 'weights/BestCOCO.pt'
opt['cfg'] = 'cfg/yolov4-csp-leaky.cfg'
opt['data'] = 'data/coco_subset.yaml'
opt['batch_size'] = 1
opt['img_size'] = 640
opt['nosave'] = False #only save final checkpoint
opt['notest']= False
opt['project'] = 'runs/test'
opt['name'] = 'test_BestFullPrecisionCOCO'
opt['precision'] = 8
opt['quantizer'] = 'faq'
opt['conf_thres'] = 0.001
opt['iou_thres'] = 0.65
opt['task'] = 'val'
opt['device'] = 'cpu' #'0'
opt['names'] = 'data/coco.names'

mosaic_labels, mosaic_predictions = test(opt, mode='unconstrained')
```

The images below demonstrate results on 3 images from the validation set. The top row displays ground truth bounding boxes (labels), and the bottom row displays the full precision network's predictions.

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Combine the two lists of images
images = mosaic_labels + mosaic_predictions

# Create a 1x3 grid of subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

# Iterate over the axes and display the images
for i, ax in enumerate(axes.flat):
    if i < len(images):
        ax.imshow(images[i])
        ax.set_axis_off()
    else:
        ax.axis('off')

# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.01, hspace=0.01)

# Display the plot
plt.show()
```

## Fine-tune an 8-bit network on MS COCO

Using the full precision network trained above as the starting point, fine-tune to optimize low precision model accuracy at 8 bits. The train function is the same as the one used for full precision training above. 

Before fine-tuning a model at low precision, the initial quantization scales for the tensors and parameters are set by using calibration. It involves running a few batches of inputs through the model to calibrate the scale factors based on data statistics. Passing <code>calibrate=True</code> to the train function below ensures that the calibration step is performed before training.

For low precision taining, NP-Pytorch mode is set to 'constrained'. In this mode, tensors and parameters, which are defined to be low precision integers, are quantized when they are cast, usually in the forward pass. The mode is set using the <code>train_mode</code> argument below.

For demonstration, the notebook trains for 2 additional epochs. In practice, it is recommended to fine-tune an 8 bit model for 60 or more epochs on a new dataset. Increasing or lowering the initial learning rate helps imporove results. When using a larger (or smaller) batch size, scale up (or down) the learning rate linearly.

For 8-bit fine-tuning, it is recommended to start from the hyper-parameters in either <code>train_coco_8bit_25epochs.</code> or <code>train_xview_8bit_60epochs_37classes_FAQ.sh</code> and modify as needed.

```python
opt = {}
opt['weights'] = 'weights/Best8bit.pt'
opt['cfg'] = 'cfg/yolov4-csp-leaky.cfg'
opt['data'] = 'data/coco.yaml'
opt['hyp'] = 'data/hyp.scratch.LR0.0000625.yaml'
opt['epochs'] = 1 
opt['batch_size'] = 1
opt['img_size'] = [640, 640]
opt['resume'] = False #resume previous training run
opt['restart'] = True #restart the most recent training, starting with epoch 0 for the lr_scheduler
opt['notest']= False
opt['workers'] =  8
opt['project'] = 'runs/train'
opt['name'] = 'expLR0.0000625_batch1_1epoch_640x640_8bit_coco'
opt['precision'] = 8
opt['quantizer'] = 'faq'
opt['conf_thres'] = 0.05
opt['img_size'].extend([opt['img_size'][-1]] * (2 - len(opt['img_size'])))  # extend to 2 sizes (train, test)
opt['save_dir'] = increment_path(Path(opt['project']) / opt['name'], exist_ok=False)  # increment run

LPlogger = logging.getLogger(__name__)
```

```python
RUN_TRAINING = False

if RUN_TRAINING:
    #Use GPU for training if available
    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

    # Load Hyperparameters from file
    with open(opt['hyp']) as f:
        hyp = yaml.load(f, Loader=yaml.FullLoader)  # load hyps

    # Train in unconstrained mode
    LPlogger.info(opt)
    results = train(hyp, opt, device, LPlogger, calibrate=True, train_mode='constrained')
```

## Measure task accuracy of trained 8-bit network on test set

Measure task accuracy of the released trained model weights, or model weights saved in the previous step on the test set below.

To measure task accuracy on a different model, change opt['weights'] to point to trained model weights below or change the test hyper-parameters for a different dataset.

```python
opt = {}
opt['weights'] = 'weights/Best8bit.pt'
opt['cfg'] = 'cfg/yolov4-csp-leaky.cfg'
opt['data'] = 'data/coco_subset.yaml'
opt['batch_size'] = 1
opt['img_size'] = 640
opt['notest']= False
opt['project'] = 'runs/test'
opt['name'] = 'test_Best8bit_coco'
opt['precision'] = 8
opt['quantizer'] = 'faq'
opt['conf_thres'] = 0.001
opt['iou_thres'] = 0.65
opt['exist_ok'] = False
opt['task'] = 'val'
opt['device'] = '0'
opt['names'] = 'data/coco.names'

mosaic_labels, mosaic_predictions = test(opt, mode='constrained')
```

The images below demonstrate results on 3 images from the test set. The top row displays ground truth bounding boxes (labels), and the bottom row displays the 8-bit network's predictions.

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Combine the two lists of images
images = mosaic_labels + mosaic_predictions

# Create a 1x3 grid of subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

# Iterate over the axes and display the images
for i, ax in enumerate(axes.flat):
    if i < len(images):
        ax.imshow(images[i])
        ax.set_axis_off()
    else:
        ax.axis('off')

# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.01, hspace=0.01)

# Display the plot
plt.show()
```

1.    During training, if there is a <code>RuntimeError</code> due to negative tensor dimensions early in training, increase the threshold value in <code>opt['conf_thres']</code>.
